Para contornar esse problema, iremos ler as imagens utilizando o Matplotlib. Acompanhem a seguinte função construída:

````python
def mostrar_imagem(imagem):
  imagem_rgb = cv.cvtColor(imagem, cv.COLOR_BGR2RGB)
  plt.imshow(imagem_rgb)
````

````python
mostrar_imagem(imagem)
````

Observe a primeira linha da função:
```
imagem_rgb = cv.cvtColor(imagem, cv.COLOR_BGR2RGB)
```
Aqui fazemos uma conversão de cor. 
O padrão de cor do OpenCV é BGR, observe o que aconteceria se apenas mostrassemos a imagem sem converter a coloração:

````python
plt.imshow(imagem)
````

A saída da imagem é "negativa", pois os canais de cor são invertidos para o padrão de leitura do Matplotlib.

Imagine a cor (255,0,0) em BGR. Essa representação é referente ao **azul no OpenCV**, mas quando passamos para o **Matplotlib temos a cor vermelha**, por se tratar de uma leitura em RGB. Por isso enxergamos a imagem com as cores invertidas.

Além disso, perceba que fazemos uma cópia da imagem para um outro padrão, isso pelo fato da imagem original (em BGR) ser últil para nosso processamento, apenas queremos vê-la em RGB para observar os resultados.

Na segunda linha da função, temos o método que mostra a imagem com o Matplotlib:


#### Pré-processando imagens...

````python
imagem_cinza = cv.cvtColor(imagem, cv.COLOR_BGR2GRAY)
````

#### Como usar o OpenCV para reconhecer faces?

O OpenCV é uma biblioteca de Visão Computacional que contém muitos artifícios para o estudo e desenvolvimento de ferramentas com tal tecnologia. Desse modo, existe nela uma série de implementações que podemos utilizar para detecção de faces.

A forma que vamos abordar aqui é utilizando o método **CascadeClassifier**.

_**E como funciona esse método?**_

O CascadeClassifier é uma implementação do **algoritmo Viola-Jones** para detecção de objetos em imagens. Esse algoritmo utiliza das **features de Haar** para identificar o que se deseja (no nosso caso, faces). 

A imagem abaixo mostra como age a extração dessas features para o reconhecimento de um rosto em uma imagem. 

<p align="center">
  <img src="https://memememememememe.me/assets/posts/training-haar-cascades/haarFace.jpg"/>
</p>

Para o melhoramento do desempenho do algoritmo, utiliza-se o **Adaboost**, um algoritmo de aprendizado de máquina utilizado para aumentar a performance de outros algoritmos.

Vamos partir para o OpenCV?



Dentro da biblioteca existe uma coleção de features de Haar que podem ser acessadas sem precisar baixar o arquivo e importá-lo. Consultando a documentação do OpenCV, assim como o repositório da biblioteca no GitHub, usaremos o arquivo **haarcascade_frontalface_alt2.xml** para identificar faces frontais.

````python
features_haar = "haarcascade_frontalface_alt2.xml"
````

Sabendo o nome do arquivo, vamos utilizar do atributo **haarcascades** do OpenCV, que nos sinaliza a localização das features de Haar dentro do conteúdo da biblioteca. Formatamos uma string com o caminho completo até o arquivo.

````python
caminho = f"{cv.haarcascades}/{features_haar}"
````

Vamos instanciar o modelo de classificação, passando as nossas features de Haar como parâmetro. 

````python
classificador = cv.CascadeClassifier(caminho)
````

Com o modelo treinado, vamos fazer as predições necessárias, utilizando o método **detectMultiScale** e passando nossa imagem em escala de cinza. Isso irá nos retornar uma matriz de coordenadas, que indica a localização dos rostos encontrados na imagem.

````python
faces = classificador.detectMultiScale(imagem_cinza)
````

Para observarmos os rostos detectados, iremos percorrer essa matriz, extraindo os pontos necessários e desenhando um retangulo em volta de cada face com o método **rectangle**.

````python
imagem_copia = np.array(imagem)
````

````python
for x,y,w,h in faces:
  cv.rectangle(imagem_copia, (x, y), (x+w, y+h), (0,255,0), 2)
````

````python
mostrar_imagem(imagem_copia)
````

#### Recortando e padronizando imagens com o OpenCV

Depois de conseguir identificar rostos com OpenCV, vamos utilizar disso para recortar e padronizar um conjunto de dados de faces. 

Quando vamos trabalhar com Machine Learning aplicado a imagens, é importante que todas estejam devidamente padronizadas, isto é, com mesmas dimensões, mesma representação de cor...

Vamos guardar as imagens cortadas numa lista

````python
imagens_cortadas = list()
````

Percorrendo mais uma vez as faces encontradas, vamos cortar as imagens utilizando um **slice**.

Logo após cortada a região onde se encontra o rosto, vamos redimensioná-la, para que todas as imagens fiquem de mesmo tamanho, para tal, utilizamos o método **resize** do OpenCV.

Feito esse processamento, vamos adicionar a imagem na nossa lista de imagens cortadas.

````python
for x,y,w,h in faces:
  face = imagem[y:y+h, x:x+w]
  face = cv.resize(face, (160,160))
  imagens_cortadas.append(face)
````

Observe abaixo que o tamanho da lista é 14, isto é, salvamos 14 faces encontradas na imagem.

````python
len(imagens_cortadas)
````

Observe também que todas as imagens possuem as mesmas dimensões

````python
for img in imagens_cortadas:
  print(img.shape)
````

Mostrando uma das imagens cortadas...

````python
mostrar_imagem(imagens_cortadas[5])
````

#### Organizando conjunto de dados em um diretório

Após padronizar os dados do nosso conjunto, vamos salvar essas imagens em um diretório, para que possamos acessá-las depois no nosso estudo.

````python
import os
````

````python
def salvar_imagens(imagens, caminho):

  if not os.path.exists(caminho):
    os.mkdir(caminho)

  index = len(os.listdir(caminho))

  for imagem in imagens:
    cv.imwrite(f"{caminho}/{index}.jpg", imagem)
    index += 1
````

````python
salvar_imagens(imagens_cortadas, "/content/drive/MyDrive/iaexpert/imagens/pessoas")
````
